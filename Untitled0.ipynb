```python
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt

def elbow_method_knn(data, column_names, min_k=1, max_k=10):
    imputer = KNNImputer()
    errors = []

    for k in range(min_k, max_k + 1):

        imput = KNNImputer(n_neighbors=k)
        imputed_data = pd.DataFrame(imput.fit_transform(data[column_names]), columns=column_names)

        variance = imputed_data.var(axis=0).mean()
        errors.append(variance)

    plt.plot(range(min_k, max_k + 1), errors, marker='o')
    plt.title('Elbow Method for KNN Imputation (Variance)')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Variance of Imputed Data')
    plt.show()

    return errors

numeric_columns = dnk.select_dtypes(include=['float64', 'int64']).columns

errors = elbow_method_knn(dnk, numeric_columns)
```

```python
import numpy as np

def find_optimal_k(errors):
    diffs = np.diff(errors)
    optimal_k = np.argmin(diffs) + 1
    return optimal_k

optimal_k = find_optimal_k(errors)
print(f"The optimal k value based on the elbow method is: {optimal_k}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore


# Assuming 'month' is the column for the month information, and pollutants are the other columns
pollutants = ['PM10', 'PM2.5', 'NOx', 'Benzene', 'Toluene', 'CO', 'SO2', 'NO', 'NO2', 'NH3', 'Xylene', 'Ozone']  # Add more pollutants as needed

# Function to detect outliers based on z-score within each month
def detect_outliers_by_month(data, pollutants, threshold=3):
    outliers_by_month = {}

    for month in dnk['month'].unique():  # Iterate through each month
        month_data = dnk[dnk['month'] == month]  # Filter data by month

        outliers_by_pollutant = {}
        for pollutant in pollutants:
            pollutant_data = month_data[pollutant]  # Get the pollutant data for the current month

            # Calculate z-scores for the pollutant
            z_scores = zscore(pollutant_data)

            # Identify outliers (z-score > threshold or z-score < -threshold)
            outliers = pollutant_data[np.abs(z_scores) > threshold]

            outliers_by_pollutant[pollutant] = outliers  # Store outliers for each pollutant

        outliers_by_month[month] = outliers_by_pollutant  # Store outliers by month

    return outliers_by_month

# Detect outliers by month
outliers = detect_outliers_by_month(dnk, pollutants)

# Display the outliers for each month and pollutant
for month, pollutants_outliers in outliers.items():
    print(f"Outliers in Month {month}:")
    for pollutant, outliers_list in pollutants_outliers.items():
        print(f"  {pollutant}: {len(outliers_list)} outliers")
    print("\n")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assuming 'df' is your DataFrame containing both pollutant and meteorological data

# Function to detect outliers (using IQR method as an example)
def detect_outliers(df, column):
    Q1 = dnk[column].quantile(0.25)
    Q3 = dnk[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dnk[(dnk[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# List of pollutants
pollutants = ['PM10', 'PM2.5', 'NOx', 'Benzene', 'Toluene', 'CO', 'SO2', 'NO', 'NO2', 'NH3', 'Xylene', 'Ozone']

# Meteorological factors
met_factors = ['AT', 'SR', 'RH', 'WD', 'RF', 'BP', 'TOT-RF']

# Dataframe to store outlier correlation results
outlier_correlation = {}

# Detect outliers and calculate correlations
for pollutant in pollutants:
    outliers = detect_outliers(dnk, pollutant)
    for met in met_factors:
        # Calculate correlation between pollutant outliers and meteorological factor
        # Using Pearson correlation for simplicity (can use other methods)
        corr, _ = pearsonr(dnk[met], dnk[pollutant])
        outlier_correlation[(pollutant, met)] = corr

# Convert correlation dictionary to a DataFrame for better visualization
outlier_corr_df = pd.DataFrame(list(outlier_correlation.items()), columns=['Pollutant-Meteorological Factor', 'Correlation'])
outlier_corr_df = outlier_corr_df.sort_values(by='Correlation', ascending=False)

# Print the correlation results
print(outlier_corr_df)

# Optional: Plot the correlations
plt.figure(figsize=(12, 8))
sns.heatmap(dnk[pollutants + met_factors].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Pollutants and Meteorological Factors')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Check stationarity for a pollutant (e.g., 'PM2.5')
def check_stationarity(series):
    result = adfuller(series.dropna(), autolag='AIC')
    print("ADF Test Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])

    if result[1] < 0.05:
        print("✅ Stationary: Reject Null Hypothesis")
    else:
        print("❌ Non-Stationary: Fail to Reject Null Hypothesis")

# Example: Check for PM2.5
check_stationarity(dnk["PM2.5"])
```

```python
import matplotlib.pyplot as plt

def plot_rolling_stats(series, window=12):
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()

    plt.figure(figsize=(10,5))
    plt.plot(series, label="Original Data", color="blue", alpha=0.5)
    plt.plot(rolling_mean, label=f"Rolling Mean ({window})", color="red")
    plt.plot(rolling_std, label=f"Rolling Std ({window})", color="green")
    plt.legend()
    plt.title("Rolling Mean & Standard Deviation")
    plt.show()

# Example: Check for Ozone
plot_rolling_stats(dnk["Ozone"])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Convert 'From Date' to datetime
dnk['From Date'] = pd.to_datetime(dnk['From Date'], format='%d-%m-%Y %H:%M')

# Define features
pollutants = ['PM10', 'PM2.5', 'NO2', 'CO', 'Ozone']
met_factors = ['AT', 'SR', 'RH', 'WD', 'RF', 'BP', 'TOT-RF']

# --- Handle Missing Values ---
dnk.dropna(subset=pollutants + met_factors, inplace=True)

# --- Normalize Data ---
scaler = StandardScaler()
scaled_datay = scaler.fit_transform(dnk[pollutants + met_factors])
df_scaled = pd.DataFrame(scaled_datay, columns=pollutants + met_factors)

# --- Feature Selection Using Regression ---
selected_features = set(pollutants)  # Always keep pollutants
for pollutant in pollutants:
    X = df_scaled[met_factors]  # Meteorological factors as predictors
    y = df_scaled[pollutant]  # Target pollutant

    # Use Random Forest Regressor
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)

    # Get feature importances
    importance = pd.Series(model.feature_importances_, index=met_factors)
    top_features = importance.nlargest(5).index.tolist()
    selected_features.update(top_features)

selected_features = list(selected_features)
print("\n✅ Selected Features Based on Regression Importance:\n", selected_features)

# --- Keep Only Selected Features ---
df_selected = dnk[selected_features]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)
# --- Visualize Feature Importance ---
plt.figure(figsize=(8, 5))
importance.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel("Feature Importance Score")
plt.ylabel("Meteorological Factors")
plt.title("Feature Importance for Pollutant Prediction")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Convert 'From Date' to datetime
dnk['From Date'] = pd.to_datetime(dnk['From Date'], format='%d-%m-%Y %H:%M')

# --- Define Features ---
pollutants = ['PM10', 'PM2.5', 'NO2', 'CO', 'Ozone']
met_factors = ['AT', 'SR', 'RH', 'WD', 'RF', 'BP', 'TOT-RF']

# --- Compute Correlation Matrix ---
corr_matrix = dnk[pollutants + met_factors].corr()

# --- Select Top 5 Highly Correlated Meteorological Factors ---
selected_factors = set(pollutants)  # Always keep pollutants
for pollutant in pollutants:
    correlated_factors = corr_matrix.loc[pollutant, met_factors].abs()
    top_5_factors = correlated_factors.nlargest(5).index.tolist()
    selected_factors.update(top_5_factors)

selected_factors = list(selected_factors)
print("\n✅ Selected Features Based on Correlation:\n", selected_factors)

# --- Keep Only Selected Features ---
df_selected = dnk[selected_factors]

# --- Normalize Data for PCA ---
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)

# --- Apply PCA (Keep 95% Variance) ---
pca = PCA(n_components=0.99)  # Retain 95% variance
pca_data = pca.fit_transform(scaled_data)

# Convert PCA result to DataFrame
pca_df = pd.DataFrame(pca_data, columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])

# --- Show Explained Variance ---
explained_variance = np.cumsum(pca.explained_variance_ratio_) * 100
print("\n📊 PCA Explained Variance (%):", explained_variance)

loadings = pd.DataFrame(pca.components_.T, index=selected_factors, columns=[f"PC{i+1}" for i in range(pca.n_components_)])
feature_importance = loadings.abs().mean(axis=1)

# Select Important Features (Threshold-based selection)
threshold = feature_importance.mean()
important_features = feature_importance[feature_importance > threshold].index.tolist()
print("Selected Important Features:", important_features)

# --- Plot Explained Variance ---
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.axhline(y=95, color='r', linestyle='--', label='95% Variance')
plt.xlabel("Number of Principal Components")
plt.ylabel("Explained Variance (%)")
plt.title("PCA Explained Variance")
plt.legend()
plt.show()
plt.figure(figsize=(8, 5))
feature_importance.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel("Average Loading Value")
plt.ylabel("Feature")
plt.title("Feature Importance Based on PCA Loadings")
plt.show()
```

```python
import torch
import numpy as np

# Function to create sequences for BiGRU
def create_bigru_sequences(data, seq_length):
    x, y = [], []
    for i in range(len(data) - seq_length):
        x.append(data[i:i + seq_length])
        y.append(data[i + seq_length])  # Predicting the next value
    return np.array(x), np.array(y)

seq_length = 24

# Create sequences BEFORE splitting (Prevents data leakage)
x_all, y_all = create_bigru_sequences(scaled_data, seq_length)

# Split the data into training and testing sets
train_size = int(len(x_all) * 0.8)
x_train, x_test = x_all[:train_size], x_all[train_size:]
y_train, y_test = y_all[:train_size], y_all[train_size:]

# Convert to PyTorch tensors using torch.from_numpy() (More efficient)
x_train_tensor = torch.from_numpy(x_train).float()
y_train_tensor = torch.from_numpy(y_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_test_tensor = torch.from_numpy(y_test).float()

# Create edge_index for GCN (Fully connected graph)
num_features = x_train.shape[-1]
edge_index = torch.tensor([[i, j] for i in range(num_features) for j in range(num_features)], dtype=torch.long).t().contiguous()

# Print shape verification
print(f"Train X Shape: {x_train_tensor.shape}")
print(f"Train Y Shape: {y_train_tensor.shape}")
print(f"Test X Shape: {x_test_tensor.shape}")
print(f"Test Y Shape: {y_test_tensor.shape}")
print(f"Edge Index Shape: {edge_index.shape}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from torch_geometric.nn import GCNConv

# Enhanced Hybrid Model: BiGRU + GCN with Attention, Dropout, Batch Norm, and Residual Connections
class Enhanced_BiGRU_GCN_Attention_Model(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, gcn_hidden_dim, dropout_prob=0.3):
        super(Enhanced_BiGRU_GCN_Attention_Model, self).__init__()

        # BiGRU for Temporal Dependency
        self.bigru = nn.GRU(input_dim, hidden_dim, num_layers=2, bidirectional=True, batch_first=True)
        self.dropout_gru = nn.Dropout(dropout_prob)

        # Attention Layer
        self.attention = nn.Linear(hidden_dim * 2, 1)

        # GCN for Spatial Dependency
        self.gcn1 = GCNConv(hidden_dim * 2, gcn_hidden_dim)
        self.bn1 = nn.BatchNorm1d(gcn_hidden_dim)

        self.gcn2 = GCNConv(gcn_hidden_dim, gcn_hidden_dim)
        self.bn2 = nn.BatchNorm1d(gcn_hidden_dim)

        # Residual Connection for GCN
        self.residual_connection = nn.Linear(hidden_dim * 2, gcn_hidden_dim)

        # Fully Connected Layer for Final Prediction
        self.fc = nn.Linear(gcn_hidden_dim, output_dim)

    def forward(self, x, edge_index):
        # BiGRU Forward Pass
        gru_out, _ = self.bigru(x)  # Output: (batch_size, seq_len, hidden_dim*2)
        gru_out = self.dropout_gru(gru_out)

        # Attention Mechanism
        attention_scores = torch.tanh(self.attention(gru_out))  # (batch_size, seq_len, 1)
        attention_weights = torch.softmax(attention_scores, dim=1)  # Normalize across time steps
        context_vector = torch.sum(attention_weights * gru_out, dim=1)  # Weighted sum over time steps

        # GCN Forward Pass with Residual Connection
        gcn_out = F.relu(self.bn1(self.gcn1(context_vector, edge_index)))
        gcn_out = F.relu(self.bn2(self.gcn2(gcn_out, edge_index)))

        # Residual Connection
        res_out = self.residual_connection(context_vector)
        gcn_out = gcn_out + res_out  # Adding Residual

        # Final Prediction
        out = self.fc(gcn_out)
        return out

# Determine device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Model Parameters
input_dim = x_train_tensor.shape[-1]   # Number of features
hidden_dim = 64
gcn_hidden_dim = 32
output_dim = y_train_tensor.shape[-1]

# Initialize Model, Loss, and Optimizer
model = Enhanced_BiGRU_GCN_Attention_Model(input_dim, hidden_dim, output_dim, gcn_hidden_dim).to(device)

# Move tensors to the device
x_train_tensor = x_train_tensor.to(device)
y_train_tensor = y_train_tensor.to(device)
x_test_tensor = x_test_tensor.to(device)
y_test_tensor = y_test_tensor.to(device)
edge_index = edge_index.to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

# Training Loop with Learning Rate Scheduler
num_epochs = 400
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    output = model(x_train_tensor, edge_index)
    loss = criterion(output, y_train_tensor)

    loss.backward()
    optimizer.step()

    # Update Learning Rate Scheduler
    scheduler.step(loss)

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Evaluation
model.eval()
with torch.no_grad():
    predictions = model(x_test_tensor, edge_index)
    test_loss = criterion(predictions, y_test_tensor)
    print(f"\n=== Test Loss: {test_loss.item():.4f} ===")

# Print shape verification
print(f"Train X Shape: {x_train_tensor.shape}")
print(f"Train Y Shape: {y_train_tensor.shape}")
print(f"Test X Shape: {x_test_tensor.shape}")
print(f"Test Y Shape: {y_test_tensor.shape}")
print(f"Edge Index Shape: {edge_index.shape}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# CPCB Breakpoints for AQI Calculation
breakpoints = {
    'PM10': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 250, 101, 200), (251, 350, 201, 300), (351, 430, 301, 400), (431, np.inf, 401, 500)],
    'PM2.5': [(0, 30, 0, 50), (31, 60, 51, 100), (61, 90, 101, 200), (91, 120, 201, 300), (121, 250, 301, 400), (251, np.inf, 401, 500)],
    'NO2': [(0, 40, 0, 50), (41, 80, 51, 100), (81, 180, 101, 200), (181, 280, 201, 300), (281, 400, 301, 400), (401, np.inf, 401, 500)],
    'CO': [(0, 1, 0, 50), (1.1, 2, 51, 100), (2.1, 10, 101, 200), (10.1, 17, 201, 300), (17.1, 34, 301, 400), (34.1, np.inf, 401, 500)],
    'Ozone': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 168, 101, 200), (169, 208, 201, 300), (209, 748, 301, 400), (749, np.inf, 401, 500)]
}

# Function to calculate sub-index for a pollutant
def calculate_sub_index(concentration, breakpoints):
    for (c_low, c_high, i_low, i_high) in breakpoints:
        if c_low <= concentration <= c_high:
            return ((i_high - i_low) / (c_high - c_low)) * (concentration - c_low) + i_low
    return np.nan

# Function to calculate AQI from multiple pollutants
def calculate_aqi(data_row):
    sub_indices = []
    for pollutant, bps in breakpoints.items():
        if pollutant in data_row:
            sub_index = calculate_sub_index(data_row[pollutant], bps)
            sub_indices.append(sub_index)
    return max(sub_indices) if sub_indices else np.nan

# Calculate AQI for Actual and Predicted Data
actual_values_2024['AQI'] = actual_values_2024.apply(calculate_aqi, axis=1)
predictions_df['AQI'] = predictions_df.apply(calculate_aqi, axis=1)
actual_values_2024['AQI'].fillna(0, inplace=True)
predictions_df['AQI'].fillna(0, inplace=True)
# Calculate Error Metrics
def calculate_metrics(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    r2 = r2_score(actual, predicted)
    return mae, mse, rmse, r2

# Hourly AQI Error Metrics
print("\n=== Hourly AQI Evaluation Metrics ===")
mae, mse, rmse, r2 = calculate_metrics(actual_values_2024['AQI'], predictions_df['AQI'])
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")

# Monthly AQI Error Metrics
print("\n=== Monthly AQI Evaluation Metrics ===")
actual_monthly_aqi = actual_values_2024['AQI'].resample('M').mean()
predicted_monthly_aqi = predictions_df['AQI'].resample('M').mean()
mae, mse, rmse, r2 = calculate_metrics(actual_monthly_aqi, predicted_monthly_aqi)
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")

# Detailed Monthly and Hourly Analysis
print("\n=== Detailed Monthly AQI Analysis ===")
monthly_analysis = pd.DataFrame({'Actual': actual_monthly_aqi, 'Predicted': predicted_monthly_aqi})
print(monthly_analysis)

print("\n=== Detailed Hourly AQI Analysis ===")
hourly_analysis = pd.DataFrame({'Actual': actual_values_2024['AQI'], 'Predicted': predictions_df['AQI']})
print(hourly_analysis.head())
```
